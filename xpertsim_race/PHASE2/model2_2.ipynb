```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuration
file_path = 'Copy of Book1_8(1).xlsx'  # Replace with your dataset path
sheet_name = 'Sheet1'  # Replace with your sheet name
target_column = "repurchase of common stock"
top_n = 6  
output_file = "top_stock_data1.csv"

# Load data
try:
    data = pd.ExcelFile(file_path)
    sheet_data = data.parse(sheet_name)
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    exit()
except ValueError as e:
    print(f"Error: {e}")
    exit()

# Reshape data
try:
    reshaped_data = sheet_data.set_index('year').transpose()
except KeyError:
    print("Error: 'year' column not found, unable to set index.")
    exit()

# Numeric conversion
reshaped_data = reshaped_data.apply(pd.to_numeric, errors='coerce')

# Check target column
if target_column not in reshaped_data.columns:
    print(f"Error: Target column '{target_column}' not found in the dataset.")
    exit()

# Calculate correlations
correlations = reshaped_data.corr()[target_column].dropna()
heatmap_data = correlations.to_frame(name="correlation").sort_values(by="correlation", ascending=False)

# Generate heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm", cbar=True)
plt.title(f"Correlation Heatmap: {target_column} vs Other Parameters")
plt.xlabel(target_column)
plt.ylabel("Other Parameters")
plt.tight_layout()
plt.show()

# Get top correlations
top_correlations = heatmap_data.head(top_n)

# Extract top parameter names
top_parameter_names = top_correlations.index.tolist()

# Include the target column
top_parameter_names.append(target_column)

# Extract data for top parameters
top_data = reshaped_data[top_parameter_names].transpose()  # Transpose to original year index
top_data = top_data.reset_index().rename(columns={"index": "Parameter"})

# Print the top_data below the chart
print("\nTop Correlated Parameter Data:")
print(top_data.to_string(index=False))  # Print DataFrame as a formatted string

# Save to CSV
top_data.to_csv(output_file, index=False)

print(f"\nData for top {top_n} correlated parameters saved to: {output_file}")
```
